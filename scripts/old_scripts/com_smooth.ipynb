```python
%matplotlib inline
import numpy as np
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
```

```python
com = np.loadtxt("./output/MW/MW_100M_beta1_vir_OM3_G4_nba_bulge_pot.txt")
```

```python
com.shape
```

```python
xcom = com[:,0]
ycom = com[:,1]
zcom = com[:,2]
rcom = np.linalg.norm(com[:,:3], axis=1)
xcom_smooth = savgol_filter(xcom, 50, 2) 
ycom_smooth = savgol_filter(ycom, 50, 2) 
zcom_smooth = savgol_filter(zcom, 50, 2)
rcom_smooth = np.sqrt(xcom_smooth**2+ycom_smooth**2+zcom_smooth**2)
```

```python
plt.plot(rcom)
plt.plot(rcom_smooth)
```

```python
vxcom = com[:,3]
vycom = com[:,4]
vzcom = com[:,5]
vcom = np.linalg.norm(com[:,3:], axis=1)
vxcom_smooth = savgol_filter(vxcom, 200, 2) 
vycom_smooth = savgol_filter(vycom, 200, 2) 
vzcom_smooth = savgol_filter(vzcom, 200, 2)
vcom_smooth = np.sqrt(vxcom_smooth**2+vycom_smooth**2+vzcom_smooth**2)
```

```python
plt.plot(vcom)
plt.plot(vcom_smooth)
```
